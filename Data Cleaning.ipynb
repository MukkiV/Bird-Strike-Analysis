```python
import pandas as pd
import numpy as np

df = pd.read_excel("DS1_C7_S4_Project_BirdStrike_Data.xlsx", sheet_name="Bird Strikes")
```

```python
# DATA CLEANING

# List of columns to fill with mode
mode_cols = [
    'Aircraft: Type',
    'Origin State',
    'Airport: Name',
    'Altitude bin',
    'Effect: Impact to flight',
    'FlightDate',
    'Aircraft: Number of engines?',
    'Aircraft: Airline/Operator',
    'When: Phase of flight',
    'Conditions: Precipitation',
    'Remarks',
    'Wildlife: Size',
    'Pilot warned of birds or wildlife?',
    'Feet above ground',
    'Is Aircraft Large?'
]

# Ensure FlightDate is datetime before filling
df['FlightDate'] = pd.to_datetime(df['FlightDate'], errors='coerce')

# Fill missing values for all columns using mode
for col in mode_cols:
    if df[col].isnull().sum() > 0:  # Fill only if missing values exist
        df[col] = df[col].fillna(df[col].mode()[0])


# Function to categorize based on 'Wildlife: Number Struck Actual'
def categorize_strike(num):
    if pd.isnull(num):
        return np.nan
    elif num == 1:
        return '1'
    elif 2 <= num <= 10:
        return '2 to 10'
    elif 11 <= num <= 100:
        return '11 to 100'
    elif num > 100:
        return 'Over 100'
    else:
        return np.nan

# Fill missing values in 'Wildlife: Number struck' using actual numeric values
df['Wildlife: Number struck'] = df['Wildlife: Number struck'].fillna(
    df['Wildlife: Number Struck Actual'].apply(categorize_strike)
)
df['Wildlife: Number struck'] = df['Wildlife: Number struck'].astype(str).str.strip()

```

```python
# Export cleaned DataFrame to CSV
df.to_csv("sales_data_cleaned.csv", index=False)

print("✅ Cleaned dataset has been exported as 'sales_data_cleaned.csv'")
```
